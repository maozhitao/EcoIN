```python
from concurrent.futures import ProcessPoolExecutor, as_completed
import time

def main(): 
    pathway_file=r'./pathway/' #Regulatory loop storage directory
    create_file(pathway_file) #Create  directory
    classification_file=r'./regulation_classification/'#Regulatory classification storage directory
    create_file(classification_file)    
    regulator_n=7 #the maximum length of the regulatory chain
    regulation_file=r'./core-EcoIN.txt' #High quality EcoIN data
    
    #File name of reactions list and metabolites list provided by users \
    #(such as Ser_Gly_Cys_met.txt and Ser_Gly_Cys_reaction.txt), \
    #which storaged in './superpathway/'
    an_name=['Ser_Gly_Cys','Phe_Tyr_Trp','Arg_Pro_Gln_Glu','Lys_Thr_Met',\
             'Val_Leu_Ile','His','Ala','Asp_Asn'] 
    #an_name=['from_3-phospho-D-glycerate']   
    with ProcessPoolExecutor() as executor:
        futures = {executor.submit(run_feed_feedforwad,eachname,regulator_n,pathway_file,regulation_file,classification_file): eachname for eachname in an_name}
        
if __name__ == '__main__':
    starttime = time.time()
    main()
    endtime = time.time()
    print('Took %.10f seconds.' % (endtime - starttime))
```
